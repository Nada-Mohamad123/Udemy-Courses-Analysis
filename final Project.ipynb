*Count Number of Courses That Are Paid and Free*

```python
plt.figure(figsize=(8, 5))

paid_count = df[df['is_paid'] == True]['course_id'].count()
free_count = df[df['is_paid'] == False]['course_id'].count()

print(paid_count, free_count)
colors=["#4CAE10", "#4CAF80"]
plt.barh(['paid','free'], [paid_count, free_count],color=colors)
plt.legend()
plt.title("Number of Paid and Free Courses", fontdict={'family':'serif', 'color':'green', 'size':20})
plt.xlabel('Course Type', fontdict={'family':'serif', 'color':'black', 'size':14})
plt.ylabel('Number of Courses', fontdict={'family':'serif', 'color':'black', 'size':14})

plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='level', y='num_lectures',data=df,color="#4CAE10")
plt.title('Number of lectures per Level',fontdict={'family':'serif', 'color':'green', 'size':20})
plt.xlabel('Level',fontdict={'family':'serif', 'color':'black', 'size':14})
plt.ylabel('Number of lectures',fontdict={'family':'serif', 'color':'black', 'size':14})
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['subject'], bins=30, kde=True, color='green')
plt.title('Distribution of Subjects', fontdict={'family':'serif', 'color':'green', 'size':20})
plt.xlabel('subject',fontdict={'family':'serif', 'color':'black', 'size':14})
plt.ylabel('Distribution',fontdict={'family':'serif', 'color':'black', 'size':14})
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(x='subject', y='num_reviews', data=df,hue='level')

plt.title('Relation between subject and number of reviews', fontdict={'family':'serif', 'color':'green', 'size':20})
plt.xlabel('Subject', fontdict={'family':'serif', 'color':'black', 'size':14})
plt.ylabel('Number of Reviews',fontdict={'family':'serif', 'color':'black', 'size':14})
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.tight_layout()
plt.title('Relation between Levels and number of Lectures', fontdict={'family':'serif', 'color':'green', 'size':20})
plt.xlabel('Level', fontdict={'family':'serif', 'color':'black', 'size':14})
plt.ylabel('Number of Lectures',fontdict={'family':'serif', 'color':'black', 'size':14})
sns.boxplot(x='level', y='num_lectures',data=df, palette='rainbow')
```

```python
df.level.value_counts().plot(kind ="pie" , autopct='%1.1f%%', colors=["#B0FBF5", "#4CAF50","#4AFA90", "#4BDBC5"], explode=(0.05, 0.02,0.1,.2), shadow = True)
plt.title("destribution of Levels", fontdict={'family':'serif', 'color':'green', 'size':20})
plt.figure(figsize=(10, 6))

plt.show()
```

```python
sns.distplot(df['num_lectures'], bins=300, kde=True,color='green')
plt.title('destribution of Number of Lectures', fontdict={'family':'serif', 'color':'green', 'size':20})
plt.xlabel('Number of Lectures', fontdict={'family':'serif', 'color':'black', 'size':14})
plt.ylabel('destribution',fontdict={'family':'serif', 'color':'black', 'size':14})
plt.figure(figsize=(10, 6))

plt.show()
```
